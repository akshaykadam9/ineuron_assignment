Q.1 Get all the cars and their types that do not qualify for clean alternative fuel vehicle

```python
non_clean_cars = df[df['Clean Alternative Fuel Vehicle (CAFV) Eligibility'] == 'No']
non_clean_cars_data = non_clean_cars[['Make', 'Model', 'Electric Vehicle Type']]
non_clean_cars_data.drop_duplicates(inplace=True)

if len(non_clean_cars_data) == 0:
    print("No cars found that do not qualify for clean alternative fuel vehicles.")
else:
    print(non_clean_cars_data)
```

Q.2 Get all TESLA cars with the model year, and model type made in Bothell City.

```python
tesla_cars_in_bothell = df[(df['Make'] == 'TESLA') & (df['City'] == 'Bothell')]
tesla_cars_data = tesla_cars_in_bothell[['Model Year']].copy()
tesla_cars_data.drop_duplicates(inplace=True)

if len(tesla_cars_data) == 0:
    print("No TESLA cars made in Bothell City found.")
else:
    print(tesla_cars_data)
```

Q.3 Get all the cars that have an electric range of more than 100, and were made after 2015

```python
cars_with_electric_range = df[(df['Electric Range'] > 100) & (df['Model Year'] > 2015)]
cars_with_electric_range_data = cars_with_electric_range[['Make', 'Model', 'Electric Range', 'Model Year']]

if len(cars_with_electric_range_data) == 0:
    print("No cars with an electric range greater than 100 and made after 2015 found.")
else:
    print(cars_with_electric_range_data)
```

4.Draw plots to show the distribution between city and electric vehicle type

```python
import matplotlib.pyplot as plt

city_ev_type_counts = df.groupby(['City', 'Electric Vehicle Type']).size().unstack()
city_ev_type_counts.plot(kind='bar', stacked=True, figsize=(40, 18))
plt.xlabel('City')
plt.ylabel('Count')
plt.title('Distribution of Electric Vehicle Types by City')
plt.legend(title='Electric Vehicle Type')
plt.show()
```
