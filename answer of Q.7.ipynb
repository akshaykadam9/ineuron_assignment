```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the CSV file into a pandas DataFrame
df = pd.read_csv("meteorite_data.csv")
```

```python
print(df.columns)
```

```python
# Convert the "year" column to datetime
df["year"] = pd.to_datetime(df["year"], format="%Y")
```

```python
# Get all Earth meteorites that fell before the year 2000
earth_meteorites_before_2000 = df[df["year"] < pd.to_datetime("2000")]
```

```python
# Plot the distribution of Earth meteorites before 2000 by mass
plt.figure(figsize=(8, 6))
plt.hist(earth_meteorites_before_2000["mass"], bins=20, color="skyblue")
plt.xlabel("Mass (g)")
plt.ylabel("Count")
plt.title("Distribution of Earth Meteorites (Before 2000)")
plt.grid(True)
plt.show()
```

```python
# Get all Earth meteorites' coordinates that fell before the year 1970
earth_meteorites_before_1970 = df[df["year"] < pd.to_datetime("1970")]
```

```python
# Plot the scatter plot of Earth meteorites' coordinates before 1970
plt.figure(figsize=(8, 6))
plt.scatter(earth_meteorites_before_1970["id"], earth_meteorites_before_1970["mass"], color="orange")
plt.xlabel("ID")
plt.ylabel("Mass (g)")
plt.title("Earth Meteorites' Mass Distribution (Before 1970)")
plt.grid(True)
plt.show()
```
